```python
#adding  items to the inventory.

ck_bts={}


n=int(input("Enter the number of item to be added to Inventory:::"))
for i in range(0,n+1):
    
    Id=int(input("Enter the  product id:"))
    name=input("Enter the name of the  product:")
    qty=2000
    typ=str("Dues-bat")
    size=input("Enter the size of the product:")
    wilo=input("enter the  willow of the product:")
    price=input("Enter the price of the product:")
    ck_bts[Id]={"Name":name,"Qty":qty,"Type":typ,"Size":size,"Willow":wilo,"price":price}

```

```python
ck_bts
```

```python
import json
bat=json.dumps(ck_bts)
fd=open("CRICKET_BAT.json","w")
fd.write(bat)
fd.close()
```
